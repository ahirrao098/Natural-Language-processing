```python
affin_scores=afinn.set_index('word')['value'].to_dict()
affin_scores
```

```python
import spacy
nlp=spacy.load('en_core_web_sm')
```

```python
sentiment_laxicon = affin_scores
```

```python
sentiment_laxicon
```

```python
def calculate_sentiment (text:str=None):
    sent_score=0
    if text:
        sentence=nlp(text)
        for word in sentence:
            sent_score += sentiment_laxicon.get(word.lemma_, 0)
            
        return sent_score
```

```python
# calculate sentiment value and add to sent_df
sent_data['sentiment_value']=sent_data['sentence'].apply(calculate_sentiment)
sent_data
```

```python
# how many words are in the sentence
sent_data['word_count']=sent_data['sentence'].str.split().apply(len)
sent_data['word_count']
```

```python
sent_data['index']=range(0,len(sent_data))
```

```python
sent_data=sent_data[['index']+['sentence']+['word_count']+['sentiment_value']]
sent_data.reset_index(drop=True)
sent_data
```

Assign labels

```python
sentiment = []
for i in range(len(sent_data)):
  c = sent_data.loc[i]['sentiment_value']
  if c>0:
    sentiment.append("positive")
  elif c<0:
    sentiment.append("Negative")
  else:
    sentiment.append("Neutral")
sent_data['Sentiment']=sentiment
sent_data
```
